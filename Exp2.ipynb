```python
#Import dependencies 
import seaborn as sns
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
df = pd.read_csv("apartments_for_rent_classified_10K.csv", sep=";", encoding='cp1252') 
df2 = df
```

```python
#Checking outliers
print(plt.figure(figsize = (10, 4), dpi = 100)) 
sns.boxplot(x = "price_display", data = df2) 
```

```python
#Visualizing by stream
plt.figure(figsize = (10, 4), dpi = 100) 
color_palette = sns.color_palette("Accent_r") 
sns.set_palette(color_palette)
sns.countplot(x = "state", data = df2) 
```

```python
#Count of internships done by students 
plt.figure(figsize = (10, 4), dpi = 100) 
color_palette = sns.color_palette("cool")
 
sns.set_palette(color_palette) 
sns.countplot(x = "bedrooms", data = df2) 
```

```python
# # #Distribution of CGPA plt.figure(figsize = (10, 4), dpi = 100) 
grp = dict(df2.groupby('bathrooms').groups) 
m = {}
for key, val in grp.items(): 
    if key in m:
        m[key] += len(val) 
    else:
        m[key] = len(val) 
        plt.title("Distribution of Bathrooms") 
        plt.pie(m.values(), labels = m.keys())
```

```python
plt.figure(figsize = (10, 4), dpi = 100)
color_palette = sns.color_palette("Accent_r") 
sns.set_palette(color_palette)
sns.countplot(x = "latitude", data = df2) 
```

```python
# Analysis of different id with time
plt.figure(figsize = (10, 4), dpi = 100) 
color_palette = sns.color_palette("magma") 
sns.set_palette(color_palette)
sns.barplot(x = "id", y = "time", data = df2) 
```

```python
# Analysis os source w.r.t. time
plt.figure(figsize = (10, 4), dpi = 100) 
color_palette = sns.color_palette("plasma") 
sns.set_palette(color_palette)
sns.barplot(x = "source", y = "time", data = df2) 
```

```python
# Smoking as per gender
df2["has_photo"].replace({"Thumbnail":0, "Photo":1}, inplace = True) 
plt.figure(figsize = (10, 4), dpi = 100)
color_palette = sns.color_palette("plasma") 
sns.set_palette(color_palette)
sns.barplot(x = "has_photo", y = "bedrooms", data = df2) 
```

```python
# Number of Photo
df2["has_photo"].replace({"Yes":1, "No":0}, inplace = True) 
plt.figure(figsize = (10, 4), dpi = 100)
color_palette = sns.color_palette("BuGn_r") 
sns.set_palette(color_palette) 
sns.countplot(x = "has_photo", data = df2)
```

```python
# Finding correlation 
numeric_df = df2.select_dtypes(include='number')
plt.figure(figsize=(10, 6))
color = sns.color_palette("magma")
sns.heatmap(numeric_df.corr(), vmax=0.9, annot=True, cmap=color)
plt.show()
```
